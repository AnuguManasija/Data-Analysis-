**Global City Population Analysis**

A dataset containing information on world cities and their populations for 2024 and 2023, along with growth rates.
 https://www.kaggle.com/code/muhammadehsan000/global-city-population-analysis/notebook



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
# Load the dataset
file_path = ('/content/World Largest Cities by Population 2024.csv')
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Calculate basic statistics
mean_population = data['Population (2024)'].mean()
median_population = data['Population (2024)'].median()
mode_population = data['Population (2024)'].mode()[0]
range_population = data['Population (2024)'].max() - data['Population (2024)'].min()
variance_population = data['Population (2024)'].var()
std_dev_population = data['Population (2024)'].std()
count_population = data['Population (2024)'].count()
percentile_population = np.percentile(data['Population (2024)'], [25, 50, 75])
iqr_population = percentile_population[2] - percentile_population[0]
# Print basic statistics
print(f"Mean: {mean_population}")
print(f"Median: {median_population}")
print(f"Mode: {mode_population}")
print(f"Range: {range_population}")
print(f"Variance: {variance_population}")
print(f"Standard Deviation: {std_dev_population}")
print(f"Count: {count_population}")
print(f"Percentiles: {percentile_population}")
print(f"IQR: {iqr_population}")
```

```python
# Summary of the dataset
print(data.describe())
```

```python
# Distribution of growth rates
plt.hist(data['Growth Rate'], bins=30,color='violet',edgecolor='black')
plt.xlabel('Growth Rate')
plt.ylabel('Number of Cities')
plt.title('Distribution of Growth Rates')
plt.show()
```

```python
# Histogram of Population (2024) and Population (2023)
sns.histplot(data['Population (2024)'], color='blue', edgecolor='black')
plt.title('Distribution of Population in 2024')
plt.xlabel('Population (2024)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(data['Population (2023)'], color='green', edgecolor='black')
plt.title('Distribution of Population in 2023')
plt.xlabel('Population (2023)')
plt.ylabel('Frequency')
plt.show()
```

.**Most cities have populations clustered around the lower end of the spectrum, with a few cities having very large populations.**

```python
# Scatter plot: Population (2024) vs Population (2023)
sns.scatterplot(x='Population (2023)',y='Population (2024)',data=data,color='purple')
plt.title('Population (2024) vs Population (2023)')
plt.xlabel('Population (2023)')
plt.ylabel('Population (2024)')
plt.show()
```

**There is a clear trend where higher-ranked cities (lower numerical rank) have larger populations. Ideally,a positive correlation if larger cities in 2023 also tend to be large in 2024.**

```python
# Pairplot for selected variables
sns.pairplot(data[['Population (2024)', 'Population (2023)', 'Growth Rate']])
plt.suptitle('Pairplot of Population and Growth Rate')
plt.show()
```

**These plots confirm the trends observed in the simpler visualizations, showing relationships between multiple variables.**

```python
# Visualize top cities by population in 2024
top_10_cities_2024 = data.nlargest(10,'Population (2024)')
sns.barplot(x='City', y='Population (2024)', data=top_10_cities_2024,color='gray')
plt.title('Top 10 Cities by Population in 2024')
plt.xlabel('City')
plt.ylabel('Population (2024)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize top cities by population in 2024
top_10_cities_2023 = data.nlargest(10,'Population (2023)')
sns.barplot(x='City', y='Population (2023)', data=top_10_cities_2023,color='pink')
plt.title('Top 10 Cities by Population in 2023')
plt.xlabel('City')
plt.ylabel('Population (2023)')
plt.xticks(rotation=45)
plt.show()
```

**The data shows significant differences in city populations.Larger populations are generally found in higher-ranked cities.There are notable regional differences in city sizes.**
